```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df
```

```python
df.drop("customerID", axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
for n in df[df["TotalCharges"]==" "].index:
    df = df.drop(n)
```

```python
df = df.reset_index(drop=True)
```

```python
df[df["TotalCharges"]==" "]
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(np.float64)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
for c in df.columns:
    print(f"{c}: {df[c].unique()}")
```
